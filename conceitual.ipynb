# O que é pensamento computacional?

Pensamento computacional é a habilidade de resolver problemas de forma lógica e estruturada, como um computador faria — dividindo o problema em partes, identificando padrões, criando regras e organizando uma solução passo a passo.
Não é sobre usar computador, e sim pensar de forma eficiente.

Exemplo simples:
Para fazer café, você segue passos: ferver água → colocar pó → coar → servir.
Isso já é pensamento computacional.

# O que é um algoritimo ? 

Algoritmo é um conjunto finito e ordenado de instruções claras e precisas, destinadas a resolver um problema"

Exemplo de algoritmo para escovar os dentes:

Pegar a escova.

Colocar pasta na escova.

Escovar os dentes por 2 minutos.

Enxaguar a boca.

Guardar a escova.



# O que é lógica de Programação?

Lógica de programação é o raciocínio usado para criar algoritmos e resolver problemas usando regras, condições e estruturas (como decisões e repetições).

Ela responde perguntas como:

Se isso acontecer, faço qual ação?

Preciso repetir algo?

Quais dados entram e quais resultados saem?
